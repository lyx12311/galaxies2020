```python
import numpy as np
import os
import astropy.units as u
import matplotlib.pyplot as plt
import HeaderP
```

```python
G=4.3009e-6
def potential(x,Ma,a):
    return -G*Ma/(x+a)

def velocity(x,Ma,a):
    return np.sqrt(x*G*Ma/(x+a)**2.)
```

```python
x=np.linspace(1,20,50)
plt.figure(figsize=(10,8))
plt.plot(x,velocity(x,0.8e12,15))
Ma,a=0.8e12,15
plt.xlabel('R [kpc]')
plt.ylabel('$v_{circ}$ [km/s]')
```

```python
def calc_eff(x,Ma,a,L):
    #L=x*np.sqrt(G*Ma*x/(x+a)**2)
    return 0.5*(L**2/x**2)+potential(x,Ma,a)
```

```python
x=8
L_circ=x*np.sqrt(G*Ma*x/(x+a)**2)
x=np.linspace(5,20,50)

plt.figure(figsize=(10,8))
plt.plot(x,calc_eff(x,0.8e12,15,L_circ))
plt.plot([8,8],[-125000,-95000],'--',label='8 kpc')
plt.xlabel('R [kpc]')
plt.ylabel('$\Phi_{eff}$')
plt.legend()
```
